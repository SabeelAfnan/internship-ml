**Research Question**

# Careful Claims

This project provides a ranked list of pages that may deserve human review based on observable search and engagement signals.

**The recommendations are intended to support decision-making and do not guarantee that refreshing a page will improve its performance.** Any observed relationships in the data (such as the correlation between search volume and clicks) should be interpreted as associations rather than proof of causation. Our goal is to offer directional insights and decision support, acknowledging that many complex, unobserved factors influence content performance.

```python
# Calculate the correlation between search volume and 90-day clicks
correlation = df['search_volume'].corr(df['clicks_90d'])
print(f"Correlation between Search Volume and 90-day Clicks: {correlation:.2f}")

# You could also visualize this if desired:
# import matplotlib.pyplot as plt
# import seaborn as sns
# plt.figure(figsize=(8, 6))
# sns.scatterplot(x='search_volume', y='clicks_90d', data=df.sample(n=1000, random_state=42))
# plt.title('Scatter plot of Search Volume vs. 90-day Clicks')
# plt.xlabel('Search Volume')
# plt.ylabel('Clicks (90 days)')
# plt.show()
```

The correlation value above shows an observed association between `search_volume` and `clicks_90d`. This highlights how we can identify relationships in the data, but it doesn't prove that increasing search volume directly 'causes' a proportional increase in clicks. Other factors are always at play, reinforcing the claim that our insights are directional and decision-support, not causal proof.
